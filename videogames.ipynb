ToDo: Describe games_with_ratings

```python
games_with_ratings['User_Score'] = games_with_ratings['User_Score'].apply(to_numeric)
games_with_ratings['User_Score'] = games_with_ratings['User_Score'] * 10
games_with_ratings.describe()
```

## How are critic scores different from user scores? Which game has the highest delta?

```python
games_with_ratings["deltaScore"] = games_with_ratings["Critic_Score"] - games_with_ratings["User_Score"]
games_with_ratings.head(15)
```

```python
games_with_ratings["deltaScore"].describe()
```

```python
len(games_with_ratings[games_with_ratings["deltaScore"] < 0])
```

```python
len(games_with_ratings[games_with_ratings["deltaScore"] > 0])
```

```python
len(games_with_ratings[games_with_ratings["deltaScore"] == 0])
```

Of the 7017 games that have a score, **3863** games have a higher user score, **2859** games have a higher critics score. In **295** games the critics and user score are equal.
